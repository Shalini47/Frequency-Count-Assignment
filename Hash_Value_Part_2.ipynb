```python
import numpy as np
```

```python
import random
import string
```

```python
# printing letters
letters = string.ascii_letters
string_list=list()
for j in range(1000):
    k=''.join(random.choice(letters) for i in range(np.random.randint(5,11,1)[0]))
    string_list.append(k)
```

```python
len(string_list[0])
```

```python
prime_no=int(input("Enter large prime number"))
```

```python
hash_value=list()
for i in range(1000):
    single_hash=0
    for j in range(len(string_list[i])):
        single_hash=single_hash+(ord(string_list[0][0])*(prime_no**(j+1)))
    hash_value.append(single_hash)
```

```python
hash_value
```

```python
String_hash_value=list(zip(hash_value,string_list))
```

```python
print("(Hash Value, String)\n",String_hash_value)
```

```python
String_hash_value.sort()
```

```python
String_hash_value
```

```python
print("Sorted Strings according to the hash values")
for i in range(1000):
    print(String_hash_value[i][1])
```
