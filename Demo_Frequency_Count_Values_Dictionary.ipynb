```python
import numpy as np
```

```python
nums=np.random.randint(1,500,1000)
freq_count=dict()
```

```python
type(nums)
freq_count.keys()
```

```python
key=list(nums)
```

```python
key
```

```python
for i in range(1000):
    if(key[i] in freq_count.keys()):
        continue
    else:
        r=key[i]
        freq_count[key[i]]=key.count(r)
    
```
