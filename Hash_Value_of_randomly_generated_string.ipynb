```python
import numpy as np
```

# Generate String Length of size between 5 to 10

```python
string_lengths=np.random.randint(low=5,high=11,size=1000)
```

```python
string_lengths[0]
```

# Create Alphabets string for small and capital letter

```python
alphabets_array= 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
alphabets_array[0]
```

# Genrate 1000 random strings of length between 5 to 10

```python
String_list=list()
for i in range(1000):
    single_string=""
    for j in range(string_lengths[i]):
        single_string=single_string+alphabets_array[np.random.randint(low=0,high=52,size=1)[0]]
    String_list.append(single_string)
```

```python
print("Strings:\n",String_list)
```

# Taking prime number from user

```python
prime_no=int(input("Enter Prime Number"))
```

# Calculate Hash Value of 1000 Random String

```python
hash_value=list()
for i in range(1000):
    single_hash=0
    for j in range(string_lengths[i]):
        single_hash=single_hash+(ord(String_list[0][0])*(prime_no**(j+1)))
    hash_value.append(single_hash)
```

```python
print("Hash Value:\n",hash_value)
```

# Zip Hash Value with its associated String

```python
String_hash_value=list(zip(hash_value,String_list))
```

```python
print("(Hash Value, String)\n",String_hash_value)
```

# Sort String_hash_Value and Reverse it to arrange in descending Order
